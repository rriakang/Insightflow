```python
import csv
import json

def csv_to_jsonl(csv_file, jsonl_file):
    
    with open(csv_file, 'r', encoding='utf-8') as csvfile, open(jsonl_file, 'w', encoding='utf-8') as jsonlfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            # 각 줄을 JSON 형식으로 변환
            json_line = {
                "prompt": f"Generate a news headline for the following content:\n\n{row['newsContent']}\n\n",
                "completion": f"{row['newsTitle']}"
            }
            jsonlfile.write(json.dumps(json_line, ensure_ascii=False) + '\n')

if __name__ == "__main__":
    # CSV 파일 경로와 JSONL 파일 경로 설정
    csv_file = "train_title_content.csv"
    jsonl_file = "train_title_content.jsonl"

    # CSV -> JSONL 변환 실행
    csv_to_jsonl(csv_file, jsonl_file)
    print(f"CSV 파일이 JSONL로 변환되었습니다: {jsonl_file}")
```
